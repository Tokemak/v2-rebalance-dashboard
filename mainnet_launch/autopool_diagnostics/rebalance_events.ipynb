```python
from mainnet_launch.constants import ALL_AUTOPOOLS, eth_client
from mainnet_launch.destination_diagnostics.fetch_destination_summary_stats import fetch_destination_summary_stats
from mainnet_launch.solver_diagnostics.fetch_rebalance_events import (
    fetch_and_clean_rebalance_between_destination_events,
    AUTOPOOL_ETH_STRATEGY_ABI,
    fetch_events,
)

dfs = {}
rebalance_dfs = {}
for autopool in ALL_AUTOPOOLS:
    strategy_contract = eth_client.eth.contract(autopool.autopool_eth_strategy_addr, abi=AUTOPOOL_ETH_STRATEGY_ABI)

    rebalance_between_destinations_df = fetch_events(strategy_contract.events.RebalanceBetweenDestinations)
    rebalance_dfs[autopool.name] = rebalance_between_destinations_df
    print(len(rebalance_between_destinations_df["block"]), autopool.name)

    uwcr_df, allocation_df, compositeReturn_out_df, total_nav_series, summary_stats_df = (
        fetch_destination_summary_stats(rebalance_between_destinations_df["block"].values, autopool)
    )
    percent_allocation_df = 100 * allocation_df.div(total_nav_series, axis=0)
    print(len(percent_allocation_df))
    dfs[autopool.name] = percent_allocation_df
    # TODO this evening
    # note this is surfacing a bigger problem, the blocks are not passe
    # allocation_df.to_csv(f'{autopool.name} allocation_after_rebalance_block.csv')
    # percent_allocation_df.to_csv(f'{autopool.name} percent_allocation_after_rebalance_block.csv')
```
