```python
import pandas as pd
import plotly.express as px
from mainnet_launch.constants import *

df = pd.read_csv('autopool_deposits_withdrawals_and_transfers.csv', low_memory=False)

stable_autopools = ['baseUSD', 'autoUSD', 'sonicUSD', 'autoDOLA']
eth_autopools = ['autoETH', 'balETH', 'autoLRT', 'baseETH', 'dineroETH']
stable_df = df[df['autopool'].isin(stable_autopools)]
eth_df = df[df['autopool'].isin(eth_autopools)]

px.box(stable_df, y=['normalized_assets'], x='autopool')
```

```python
transfers = df[df['event'] == 'Transfer'].copy()[['hash', 'from', 'to', 'normalized_value']]
not_transfers = df[df['event'] != 'Transfer'].copy()[['event','hash', 'autopool', 'timestamp', 'normalized_assets' ]]


mints = transfers[transfers['from'].str.startswith('0x0000000000000000000000000000000000000000')].copy()
burns = transfers[transfers['to'].str.startswith('0x0000000000000000000000000000000000000000')].copy()
autoUSD_withdraws = not_transfers[(not_transfers['event'] == 'Withdraw') & stable_df['autopool'].isin(['autoUSD'])].copy()
autoUSD_deposits = not_transfers[(not_transfers['event'] == 'Deposit') & stable_df['autopool'].isin(['autoUSD'])].copy()
autoUSD_transfers = not_transfers[(not_transfers['event'] == 'Transfer') & stable_df['autopool'].isin(['autoUSD'])].copy()
```

```python
buckets = [(0, 100),(100, 1000), (1000, 10000), (10000, 100000), (100000, 1000000), (1_000_000, 10_000_000)]
# add a column called "bucket" to autoUSD_deposits that categorizes the normalized_assets into the buckets
def categorize(value):
    for i, (low, high) in enumerate(buckets):
        if low <= value < high:
            return f'{low}-{high}'
    return 'other'
autoUSD_deposits['bucket'] = autoUSD_deposits['normalized_assets'].apply(categorize)

agg_df = (
    autoUSD_deposits
    .groupby('bucket')['normalized_assets']
    .agg(['count', 'mean','median','sum'])
    .round(2)
    .sort_values(by='sum', ascending=False)
)

# 2) Format the dollar columns
for col in ['mean','median','sum']:
    agg_df[col] = agg_df[col].apply(lambda x: f'${x:,.2f}')

# 3) (Optional) add thousands separators to your counts, too
agg_df['count'] = agg_df['count'].apply(lambda x: f'{x:,}')

```
