```python
def _extract_top_level_autopool_data(autopool_info:dict, decimals:int) -> dict:
    return {
        'autopool':autopool_info['symbol'],
        'totalSupply':autopool_info['totalSupply'] / 1e18,
        'totalIdle':autopool_info['totalIdle'] / 10 ** decimals,
        'totalDebt':autopool_info['totalDebt'] / 10 ** decimals,
        'navPerShare':autopool_info['navPerShare'] / 10 ** decimals,
    }


def _extract_destination_info(destinations_info_list: list[dict], decimals:int) ->list[dict]:
    clean_destination_info_list = []
    for destination_info in destinations_info_list:
        clean_destination_info = {
            'destinationVaultAddress': ETH_CHAIN.client.toChecksumAddress(destination_info['vaultAddress']),
            'autoPoolOwnsShares':destination_info['autoPoolOwnsShares'] / 1e18, 
            'actualLPTotalSupply':destination_info['actualLPTotalSupply'] / 1e18,
            'autoPoolOwnsShares':destination_info['autoPoolOwnsShares'] / 1e18,
            'debtValueHeldByVault':destination_info['debtValueHeldByVault'] / 10 ** decimals,
            'underlyingTokenSymbols': [t['symbol'] for t in destination_info['underlyingTokenSymbols']]
        }

        clean_destination_info_list.append(clean_destination_info)
    
    return clean_destination_info_list

def _extract_top_level_info_from_lens_contract(autopool_dest_tuple:tuple[dict,list[dict]]) -> dict:
    autopool_info, destinations_info_list = autopool_dest_tuple
    if autopool_info['baseAsset'].lower() in [WETH(chain).lower() for chain in ALL_CHAINS]:
        decimals = 18
    elif autopool_info['baseAsset'].lower() ==  [USDC(chain).lower() for chain in ALL_CHAINS]:
        decimals = 6
    else:
        raise ValueError(f'Unexpected {autopool_info['baseAsset']=}')
    
    clean_autopool_info = _extract_top_level_autopool_data(autopool_info, decimals)
    clean_destination_info_list = _extract_destination_info(destinations_info_list, decimals)

    return {
        'clean_autopool_info':clean_autopool_info,
        'clean_destination_info_list': clean_destination_info_list,
    }

def process_getPoolsAndDestinations_response(getPoolsAndDestinations:dict) -> dict:
    """
    Returns a dictionary that has

    For the autopool
    total supply, total idle, total debt, and nav per share

    for each destination in the list of destinations

    """
    autopool_dest_tuples = [
        (a, d) for a, d in zip(getPoolsAndDestinations["autopools"], getPoolsAndDestinations["destinations"])
    ]
    by_autopool_dict = {}
    for autopool_dest_tuple in autopool_dest_tuples:
        cleaned_reponse = _extract_top_level_info_from_lens_contract(autopool_dest_tuple)
        by_autopool_dict[cleaned_reponse['clean_autopool_info']['autopool']] = cleaned_reponse

    return by_autopool_dict

```
