```python
from mainnet_launch.database.schema.full import DestinationStates, Destinations
from mainnet_launch.database.postgres_operations import get_full_table_as_df_with_block,get_full_table_as_df
from mainnet_launch.constants import BASE_CHAIN, ARBITRUM_CHAIN
import plotly.express as px

destination_vault_address = '0x945a4f719018edBa445ca67bDa43663C815835Ad'
df = get_full_table_as_df_with_block(DestinationStates, where_clause=DestinationStates.chain_id == ARBITRUM_CHAIN.chain_id)
destinations = get_full_table_as_df(Destinations)
df['name'] = df['destination_vault_address'].map(destinations.set_index('destination_vault_address')['name'])
df
```

```python
import plotly.express as px
import plotly.io as pio

pio.templates.default = None
# Resample to get last value of each day
# Resample to get last value of each day
df_daily = df.resample('D').last()

# Calculate 30-day rolling APY based on price change
df_daily['apy'] = 100 * ((df_daily['lp_token_safe_price'] / df_daily['lp_token_safe_price'].shift(30)) ** (365/30) - 1)

# Plot the APY
px.scatter(df_daily, y='apy', color='name', title='LP Token APY (30-Day Rolling)')
```
